```python
def func(x):
    return x**2

def plot_function(func):
    #linspace(start , end , no.of values to construct)
    x_values = np.linspace(-10,10,1000)
    y_values  = [func(x) for x in x_values] 
    plt.plot(x_values,y_values)

plot_function(func)
plot_function(lambda x: np.sin(x))
plot_function(lambda x: x**2+4*x+3)
    
#lambda function are the anonymus functions with out a name and the defination of that function written in one line
```

```python
def plot_derivative(func):
    x_values = np.linspace(-5,5,1000)
    
    delta_x = 0.0001
    y_values = (func(x_values+delta_x) - func(x_values))/delta_x #(f(x + ∆x) - f(x))/∆x
    plt.plot(x_values,y_values)

plot_derivative(lambda x:np.sin(x))
```

```python
def f(x):
    return x**2+5*x
x = np.linspace(-5,5,1000)
dfdx = np.gradient(f(x),x)
plt.plot(x,f(x),label = "f(x)")
plt.plot(x,dfdx,label = "df/dx")
plt.legend()
plt.show()
```
